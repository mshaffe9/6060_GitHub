# Pre-Processing

**Handle missing values,
encoding,
IDness,
handle outliers,
normalization and scaling,
stability and health checks**

```python
pd.set_option('display.max_rows', None)
```

```python
df.isnull().sum()
```

```python
df1 = df.copy()
```

```python
#Delete columns Alley PoolQC, Fence, MiscFeature because they had most values missing and/or 
#weren't deemed as important to predict listing price
df1.drop(columns=['Alley','PoolQC','Fence','MiscFeature','LotFrontage','FireplaceQu'], inplace=True)
```

```python
df1.isnull().sum()
```

```python
#dropped the rest of the NA values because there were not many compared to the entire dataset
#and the values within these columns not deemed important to predicting list price
df1.dropna(inplace=True)
df1.isnull().sum()
```

**Drop Duplicates**

```python
df1 = df1.drop_duplicates()
df1.shape
```

```python
# encode categorical variables using one-hot encoding
df1 = pd.get_dummies(df1, columns=['Street', 'CentralAir', 'PavedDrive'], prefix=['Street', 'CentralAir', 'PavedDrive'])
```

**Outliers**

```python
from scipy.stats import zscore
df1['zscore_SalePrice'] = zscore(df1['SalePrice'])
```

```python
threshold = 2
df1['outliers'] = np.where((df1['zscore_SalePrice'] - threshold > 0), True, np.where(df1['zscore_SalePrice'] + threshold < 0, True, False))
```

```python
df1.drop(df1[df1['outliers'] == True].index,inplace=True)
```

**Standardization**

```python
# standardize the values of a numeric column
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df1['LotArea'] = scaler.fit_transform(df1[['LotArea']])
```

```python
# standardize the values of a numeric column
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df1['LotArea'] = scaler.fit_transform(df1[['MSSubClass']])
```

**Use StandardScaler to have distribution mean value 0 and standard deviation of 1. This will help with the outliers identified previously.**

```python
# calculate summary statistics for a numeric column
print(df1['1stFlrSF'].describe())
```

```python
# calculate summary statistics for a numeric column
print(df1['2ndFlrSF'].describe())
```

```python
# apply a robust scaling to a numeric column
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
df1['1stFlrSF'] = scaler.fit_transform(df1[['1stFlrSF']])
```

```python
# apply a robust scaling to a numeric column
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
df1['2ndFlrSF'] = scaler.fit_transform(df1[['2ndFlrSF']])
```

```python
# calculate the skewness and kurtosis of a numeric column
print(df1['SalePrice'].skew())
print(df1['SalePrice'].kurtosis())
```

```python
# perform a t-test to compare the means of two numeric columns
from scipy.stats import ttest_ind

t, p = ttest_ind(df1['SalePrice'], df1['1stFlrSF'])
print(t, p)
```

```python
# perform an ANOVA test to compare the means of two or more numeric columns
from scipy.stats import f_oneway

f, p = f_oneway(df1['SalePrice'], df1['1stFlrSF'])
print(f, p)
```

```python
# calculate the correlation between two numeric columns
print(df1['SalePrice'].corr(df1['1stFlrSF']))
```

```python
# create a heatmap to visualize the correlation between multiple columns
plt.figure(figsize=(30, 20))
sns.heatmap(df1.corr(), cmap='RdYlGn', annot=True)
plt.show()
```

There may be collinearity in features that are highly correlated (above 0.6 or below -0.6) to one another. 
Besides SalePrice (because this is the dependent variable we are trying to predict), these features include OverallQual, TotalBsmtSF, 1stFlrSF, GrLivArea, GarageCars, GarageArea, Street_Grvl, Street_Pave, CentralAir_N, CentralAir_Y, PavedDrive_N, PavedDrive_Y, YearBuilt, YearRemodAdd, TotRmsAbvGrd, 2ndFlrSF, BedroomAbvGrd, HalfBath, FullBath.
